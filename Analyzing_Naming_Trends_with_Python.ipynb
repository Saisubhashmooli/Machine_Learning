```python
import os
import pandas as pd 
import numpy as np
import zipfile 
import matplotlib.pyplot as plt
```

```python
zf = zipfile.ZipFile('E:/dataset/National_names.zip') 
names = pd.read_csv(zf.open(zipfile.ZipFile.namelist(zf)[0])) 
```

```python
names.head()
```

```python
names = names.rename(columns={'Mary':"Names", 'F':"Gender", '7065':"Birth"})
names.head()
```

```python
names.info()
```

```python
names.shape
```

```python
by_gender = names.groupby('Gender')
by_gender.size().plot(kind='bar')
```

```python
names.sort_values(by='Birth', ascending=False)
```

```python
names.sort_values(by='Birth', ascending=False).head(100)
```

```python
names.groupby('Names')['Birth'].mean().sort_values(ascending=False).head(100)
```

```python
names.groupby('Names')['Birth'].count().sort_values(ascending=False).head(100)
```

```python
counts = pd.DataFrame(names.groupby('Names')['Birth'].mean())  
  
counts['Top'] = pd.DataFrame(names.groupby('Names')['Birth'].count()) 
  
counts.head() 
```

```python
counts.sort_values('Top', ascending = False).head(10)
```
